```python
import pandas as pd

# Load your dataset
weather_df = pd.read_csv('Forcasted Weather data.csv')

```

```python
# Convert the 'ts' column to datetime
weather_df['datetime'] = pd.to_datetime(weather_df['ts'])

# Set the 'ts' column as the index
weather_df.set_index('datetime', inplace=True)

# Create a complete hourly range of timestamps from midnight on Jan 1, 2022
full_index = pd.date_range(start='2022-01-01 00:00', end=weather_df.index[-1], freq='H')

# Reindex the DataFrame to include all timestamps in the range
full_weather_df = weather_df.reindex(full_index)

# Backfill the missing data with the earliest available data (from Jan 2, 2022, 9:00)
full_weather_df = full_weather_df.bfill(limit=33)

# Display the rows from Jan 1 to Jan 2, 2022, to verify
#print(full_weather_df.loc['2022-01-01':'2022-01-02 09:00'])
full_weather_df=full_weather_df.reset_index()
full_weather_df=full_weather_df.drop('ts',axis=1)
full_weather_df
full_weather_df.rename(columns={'index':'ts'})
weather_df.to_csv('Forcasted Weather data new.csv', index=False)
```
